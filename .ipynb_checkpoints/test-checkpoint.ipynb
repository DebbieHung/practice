```python
url="https://www.hpa.gov.tw/Pages/List.aspx?nodeid=4086"
```

```python
chrome=get_chrome(url)
chrome
```

```python
soup=BeautifulSoup(chrome.page_source,'lxml')
```

```python
allText=soup.find("div",class_="htmlBlock").text.split("\n\n\n\n\n\xa0")
allText
```

```python
len(allText)
```

```python
allText[0].replace("\n",'').replace(" ",'').replace("\xa0","")
```

```python
for i in range(len(allText)):
    allText[i]=allText[i].replace(" ",'').replace("\xa0","")
print(allText)
```

```python
allText[0].split('\n')[0]
```

```python
kind=[]
for i in range(len(allText)):
    kind.append(allText[i].split('\n')[0])
print(kind)
```

```python
allText[0].replace("\n",'').replace(kind[0],'')
```

```python
desc=[]
for i in range(len(allText)):
    desc.append(allText[i].replace("\n",'').replace(kind[i],''))
print(desc)
```

```python
dict={}
for i in range(len(kind)):
    dict[kind[i]]=desc[i]
print(dict)
```

```python
dict={}
dictall={}
data=[]
for i in range(len(kind)):
    for j in range(len(kind)):
        if j == 5:
            dict["value"] = 0.167
            dict["name"] = kind[5]
            dictall['chart']=dict
            dictall['info']=desc[5]
        else:
            dict["value"] = 0.167
            dict["name"] = kind[i - 1]
            dictall['chart']=dict
            dictall['info']=desc[i-1]
            dict = {}
            dictall={}

    data.append(dictall)
print(data)
```

```python
for i in range(len(data)):
    print(data[i]['info'])
    print('---')
```

```python
for i in range(len(data)):
    print(data[i]['chart'])
```

```python
chart=[data[i]['chart'] for i in range(len(data))]
print(chart)
```
